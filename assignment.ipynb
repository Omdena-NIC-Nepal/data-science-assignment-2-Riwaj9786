# Task 7
- Merging and Joining DataFrames

```python
df1 = pd.DataFrame({'C': ['C0', 'C1', 'C2'],
                    'D': ['D0', 'D1', 'D2']},
                    index = ['K0', 'K1', 'K2'])
df1
```

```python
df2 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                   'B': ['B1', 'B2', 'B3']}, 
                  index = ['K0', 'K1', 'K2'])
df2
```

```python
join_df = df1.join(df2)
join_df
```

```python
data_first = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}
df1 = pd.DataFrame(data_first)
```

```python
data_second = {
    'ID': [3, 4, 5, 6],
    'Email': ['charlie@example.com', 'david@example.com', 'eve@example.com', 'frank@example.com'],
    'Salary': [70000, 80000, 90000, 100000]
}
df2 = pd.DataFrame(data_second)
```

```python
#Merging the DataFrame on the common column 'ID'
merged_df = pd.merge(df1, df2, on='ID', how='inner')
```

```python
merged_df
```

```python
merged_df2 = pd.merge(df1, df2, on='ID', how='left')
merged_df2
```
